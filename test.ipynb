```python
import yfinance as yf
import matplotlib.pyplot as plt # should be plotext here for the terminal
```

```python
gme = yf.Ticker("GME")

gmeH = gme.history(period="1d", interval="1m")
gmeH["Avg"] = (gmeH["Open"]+gmeH["Close"])/2
gmeH.head()
gmeH.plot(y="Avg")
```

```python
x = gmeH.index.tolist()
y = gmeH["Avg"].tolist()

plt.plot(x,y)
plt.show()

plt.scatter(x,y)
```
